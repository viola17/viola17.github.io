```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import date, datetime, timedelta
```

## Clean Data

```python
bulk = pd.read_csv("Bulk.csv")
weather = pd. read_csv("Weather.csv")

bulk.isna().any()
weather.isna().any()

bulk.duplicated().any()
weather.duplicated().any()
```

```python
# create dataframe with 47 time spot on Oct 13

curr = datetime(2019, 10, 13,0,30)
end = datetime(2019, 10, 13,23,30)
delta=  timedelta(**{'minutes':30})
dates = [pd.to_datetime(curr)]
while curr < end:
    curr += delta
    dates.append(pd.to_datetime(curr))
    
temp = pd.DataFrame({"DateTime": dates, 
                    "Power (kW)":[np.nan for i in range(47)]}) 

# complete power consumption dataset with additional time spots 
bulk = bulk.append(temp,ignore_index=True)

# glance at the bulk data and see how many missing values are there
bulk.info()
bulk
```

```python
# Take a look at where missing values occurred

bulk[bulk.isna().any(axis=1)].head()
```

```python
# in order to estimate the missing values, plot part of the data and see if any trend can be detected

bulk['Power (kW)'][5000:5200].plot(linewidth=0.5)
```

```python
#bulk = bulk.interpolate() # since we are gonna drop the na values in the following step, we abandoned interpolation 
```

```python
# transform the data type to date time 

weather['DateTime']=pd.to_datetime(weather[['Year','Month','Day','Hour','Minute']])
bulk['DateTime'] = pd.to_datetime(bulk.DateTime)

right_df = bulk.merge(weather,how='right', left_on='DateTime', right_on='DateTime')
left_df = bulk.merge(weather,how='left', left_on='DateTime', right_on='DateTime')

left_df.head(10)
```

```python
right_df.head(10)
```

```python
# Merge datasets 
right_df.set_index('DateTime')
left_df.set_index("DateTime")
full_df = pd.concat([left_df,right_df], axis=0).drop_duplicates()

# interpolate weather conditions
full_df[full_df.columns[7::]] = full_df[full_df.columns[7::]].interpolate().bfill()
full_df['Year']=full_df['DateTime'].dt.year
full_df['Month']=full_df['DateTime'].dt.month
full_df['Day']=full_df['DateTime'].dt.day
full_df['Hour']=full_df['DateTime'].dt.hour
full_df['Minute']=full_df['DateTime'].dt.minute

# set datetime as index 
full_df = full_df.set_index('DateTime').sort_index()

# final dataset
full_df
```

```python
full_df.info() # 47 missing values in the column power left for prediction 
```

```python
# create dataframe with new columns of lagged values
lagged_df = full_df.copy()
lagged_df = lagged_df[0:17425] # 17425 entries of collected power data 

# Next day's load values.
n=48
lagged_df['load_tomorrow'] = lagged_df['Power (kW)'].shift(-n)   
# Get the data of last week
for day in range(8):
    lagged_df['temperature_d' + str(day)] = lagged_df['Temperature'].shift(n*day)
    lagged_df['wind_speed_d' + str(day)] = lagged_df['Wind Speed'].shift(n*day)
    lagged_df['humidity_d' + str(day)] = lagged_df['Relative Humidity'].shift(n**day)
    lagged_df['precipitation_log_d' + str(day)] = lagged_df['Total Precipitation'].shift(n*day)
    lagged_df['wind_direction'+ str(day)] = lagged_df['Wind Direction'].shift(n*day)
    
    lagged_df['load_d' + str(day)] = lagged_df['Power (kW)'].shift(n*day)


lagged_df = lagged_df.dropna()

lagged_df = lagged_df.drop(columns=['Temperature', 'Wind Speed', 'Relative Humidity',
                                    'Total Precipitation','Wind Direction','Power (kW)'])

lagged_df
```

```python
# plot predictive and actual results on the same graph
def subplot_prediction(actual, prediction,prediction_label):
    fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(20, 15))
    
    con_df = pd.concat([actual.rename('Actual'),pd.DataFrame(prediction, index=actual.index, columns=[prediction_label])], axis=1)
    axes[0].set_title('Actual vs Prediction - One day')
    axes[0].set_ylabel('Power(MW)')
    axes[0].set_xlabel('Datetime')
    con_df.plot(ax=axes[0])
    axes[0].set_xlim(left=con_df.index[-24*1] , right=con_df.index[-1])
    
    axes[1].set_title('Actual vs Prediction - One week')
    axes[1].set_ylabel('Power(MW)')
    axes[1].set_xlabel('Datetime')
    con_df.plot(ax=axes[1])
    axes[1].set_xlim(left=actual.index[-24*7] , right=actual.index[-1])
    
    axes[2].set_title('Actual vs Prediction - One month')
    axes[2].set_ylabel('Power(MW)')
    axes[2].set_xlabel('Datetime')
    con_df.plot(ax=axes[2])
    axes[2].set_xlim(left=actual.index[-24*7*4] , right=actual.index[-1])
    
    plt.tight_layout()
    plt.show()
    
# show importance of variables (if any)
def importance(reg,X_train, title):
    feat_imp = pd.DataFrame({'importance':reg.feature_importances_})    
    feat_imp['feature'] = X_train.columns
    feat_imp.sort_values(by='importance', ascending=False, inplace=True)
    feat_imp.sort_values(by='importance', inplace=True)
    feat_imp = feat_imp.set_index('feature', drop=True)
    feat_imp.plot.barh(title=title, figsize=(10,18))
    plt.xlabel('Feature Importance Score')
```

```python
# split test and train data and define X, y values 

X = lagged_df.drop(columns=['load_tomorrow'])
y = lagged_df['load_tomorrow']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=False)
```

## Other models that worth a try 

```python
def Regressor(data):
    X_df = data.drop(columns=['load_tomorrow'])
    Y_df = data['load_tomorrow']
    X = X_df.values
    Y = Y_df.values
    test_size = 0.3
    seed = 61801
    X_train, X_validate, Y_train, Y_validate = train_test_split(X, Y, test_size = test_size, random_state = seed)
    num_folds = 10
    scoring = 'explained_variance'
    models = []
    models.append(('KNN', KNeighborsRegressor()))
    models.append(('DT', DecisionTreeRegressor()))
    models.append(('RF', RandomForestRegressor()))
    
    #cross validation and RSS results
    results = []
    names = []
    preds = []
    for name, model in models:
        kfold = KFold(n_splits=num_folds, random_state=seed,shuffle = False)
        cv_results = cross_val_score(model, X_train, Y_train, cv=kfold, scoring=scoring)
        results.append(cv_results)
        names.append(name)
        
        model.fit(X_train, Y_train)
        prediction = model.predict(X_validate)
        preds.append((name,prediction))
        
        rss = np.sum((Y_validate - prediction)**2)
        
        msg = "%s: %f (%f) RSS:%f" % (name, cv_results.mean(), cv_results.std(),rss)
        print(msg)
        
    for n,p in preds:
        subplot_prediction(y_test, p ,prediction_label= n)
    
```

# Conclusion

## Approach and Evaluation

### Predictors : 
To predict the power consumption for the upcoming day, I used 7 lagged values of weather records (temperature/ humidity etc)  as well as the power consumption in the preceding week. 

### Data cleaning: 
For version A, I combined the two datasets by date time, and used interpolation to estimate the missing weather conditions for time spots at every half hour.  (Weather may change dramatically during an hour, but it is left to think of a better way to estimate missing values. For Version B, I only used actual data to predict 24 power consumption data. ) After creating columns of lagged and the current values, I dropped all rows with NaN and duplicated columns, then split the remaining data into train and test sets. 

### Model choices:
Linear regression, Ada boost, Bagging regression, and extra trees were the first that came to mind. Then I also tried other regular regressors such as decision tree, k-nearest neighbours, and random forest. 

To compare these models, I used 5-fold cross-validation to obtain mean accuracy scores and calculated RSS to measure how well they performed , then plotted actual and predicted power consumptions on the same graphs respectively. The best model should receive the highest accuracy score, lowest RSS and predicted as close as to the actual power consumption. 

Though it turned out that random forest gave the best accuracy score and lowest RSS, the prediction plot was not quite in line with the actual values, since the spikes in its prediction plot often pointed in the opposite direction to the actual consumption. This should be what we try to avoid in predicting power usage. 

Therefore, per the aforementioned principle, I chose Extra trees to be the best fit model for ultimate prediction. Compared to other models, Extra trees gave predictions closest to the actual values. Bagging or linear regression could be our second choice because it performed equally well except it had a lower accuracy score and higher RSS in comparison to Extra trees. 

### Important Features
Hours have the highest correlation with the power consumption. It seems the weather of the day does not have much to do with power usage. From the Ada boost model, the three most important variables are power usage of the day, power usage on the same day last week, and hour. Either from the correlation matrix or from the variable importance plot, precipitation is the least correlated with power consumption. To test if subset of data performs better, I fit another Ada Boost model with only 4 predictors. The result shows that the accuracy is relatively higher, while RSS is relatively lower with a subset of  important variables only.
